```python
#category wise sales 

import matplotlib.pyplot as plt

category_sales = df.groupby('PRODUCTLINE')['SALES'].sum()

plt.figure(figsize=(10,5))
category_sales.plot(kind='bar')

plt.title('Category-wise Sales')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))

category_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Category-wise Sales')
plt.ylabel('')
plt.show()
```

```python
region_sales = df.groupby('TERRITORY')['SALES'].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')

plt.title('Region-wise Sales')
plt.xlabel('Territory')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(8,8))

region_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Region-wise Sales')
plt.ylabel('')
plt.show()
```

```python
top_products = (
    df.groupby('PRODUCTLINE')['SALES']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print(top_products)
```

```python
 

top_products.plot(kind='bar', figsize=(8,5))

plt.title("Top 5 Performing Products")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
top_regions = (
    df.groupby('TERRITORY')['SALES']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print(top_regions)

top_regions.plot(kind='bar', figsize=(8,5))

plt.title("Top Performing Regions")
plt.xlabel("Territory")
plt.ylabel("Total Sales")
plt.show()

```
